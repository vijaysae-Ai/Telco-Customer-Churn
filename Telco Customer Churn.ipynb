```python
import pandas as pd
```

```python
df=pd.read_csv('WA_Fn-UseC_-Telco-Customer-Churn.xls')
```

```python
df.head()
```

```python
##feature Selection

X=pd.get_dummies(df.drop(['customerID','Churn'],axis=1))
y=df['Churn'].apply(lambda x: 1 if x=='Yes' else 0 )
```

```python
X
```

```python
y
```

```python
#train test split
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
X_train.head()
```

```python
X_test.head()
```

```python
from tensorflow.keras.models import Sequential,load_model
from tensorflow.keras.layers import Dense
from sklearn.metrics import accuracy_score
```

```python
#model
model =Sequential()
model.add(Dense(units=30,activation='relu',input_dim=len(X_train.columns)))
model.add(Dense(units=60,activation='relu'))
model.add(Dense(units=1,activation='sigmoid'))
```

```python
model.compile(optimizer='sgd',loss='binary_crossentropy',metrics='accuracy')
```

```python
model.fit(X_train,y_train,batch_size=32,epochs=100)
```

```python
y_pred=model.predict(X_test)
```

```python
y_pred=[0 if val<0.5 else 1 for val in y_pred]
```

```python
accuracy_score(y_test,y_pred)
```

```python
model.save('fmodel')
```
